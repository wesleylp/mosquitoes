## Choosing video to test

```python
video_path = '../data/DJI4_cam/'
video_name = 'DJI_0001.MOV'
video = os.path.join(video_path, video_name)
```

## Creating video object

```python
vid = video_utils.videoObj(videopath=video)
```

### We can access specific video information:

```python
total_frames = vid.videoInfo.getNumberOfFrames()
frame_rate = vid.videoInfo.getFrameRate()
print('This video has {} frames.'.format(total_frames))
print('Frame rate: {}.'.format(frame_rate))
```

### We can print all available video information

```python
vid.videoInfo.printAllInformation()
```

## Getting a certain video frame

```python
#using method get_frame to capture a certain frame number
ret, frame, frame_size = vid.get_frame(frame_req=20)
```

```python
#displaying frame
plt.imshow(cv2.cvtColor(frame,cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
print('image size: ', frame_size)
```
