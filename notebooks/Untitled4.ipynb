```python
import os
import glob
import pickle

import matplotlib.pyplot as plt
import pandas as pd
```

```python
list_files = glob.glob("/home/wesley.passos/Public/phase_correlation/*.pkl")
```

```python
for file in list_files:
    with open(file, 'rb') as f:
        df = pickle.load(f)
    
    print(file)
    
    plt.figure(figsize=(16,6))    
    
    plt.subplot(3,2,1)
    plt.hist(df.conf, bins=20)
    plt.xlim([0,1])        
    
    plt.subplot(3,2,2)
    plt.plot(df.conf, '.')
    plt.ylim([0,1])    
    plt.xlabel('frame idx')
    plt.ylabel('corr')
    
    plt.subplot(3,2,4)
    plt.plot(df.x, '.')
    plt.xlabel('frame idx')
    plt.ylabel('x displacement')
    
    plt.subplot(3,2,6)
    plt.plot(df.y, '.')
    plt.xlabel('frame idx')
    plt.ylabel('y displacement')
    
    plt.show()
```
